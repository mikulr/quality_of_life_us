```python
from bs4 import BeautifulSoup as bs
import pandas as pd
import re
from selenium import webdriver
from bs4 import BeautifulSoup
from webdriver_manager.chrome import ChromeDriverManager
```

```python
driver = webdriver.Chrome(ChromeDriverManager().install())
driver.get('https://www.usnews.com/news/best-states/rankings/opportunity/affordability')
```

```python
soup = BeautifulSoup(driver.page_source, 'lxml')
tables = soup.find_all('table')

# print(soup.prettify())
```

```python
#the return is a jumbled array of tables, what I want is in the last one
data = pd.read_html(str(tables))
data
```

```python
#create affordability dataframe from scraped table
aff_df=pd.DataFrame(data[102])
#drop garbage columns, NaN rows and reset index
aff_df=aff_df.drop(columns={'Unnamed: 1', 'Unnamed: 3', 'Unnamed: 5'}).dropna().reset_index(drop=True)
aff_df.head()
```

```python
#regex def to pass state name through to split doubling of state name 
def split_it(state):
    return re.findall(r'[A-Z][^A-Z]*', state)
```

```python
#apply split it which creates nested list in column
aff_df['State']=aff_df['State'].apply(split_it)
```

```python
#Add new column to add the corrected state name 
aff_df['state']=''
```

```python
#add correct state name in column 
#if it is a two name state (eg, North Caroilina) the split list will be four list items
#will have to concat the words together
for x in range(0, 50):
    if len(aff_df['State'][x]) == 4:
        aff_df['state'][x] = aff_df['State'][x][0] + aff_df['State'][x][1]
    else:
        aff_df['state'][x] = aff_df['State'][x][0]
```

```python
aff_df['Overall_Affordability_Rank']=aff_df['Rank'].astype(int)
aff_df['Cost_of_Living_Rank']=aff_df['Cost of Living'].astype(int)
aff_df['Housing_Affordability_Rank']=aff_df['Housing Affordability'].astype(int)
aff_df=aff_df.drop(columns={'Rank','Cost of Living', 'Housing Affordability','State'})
```
