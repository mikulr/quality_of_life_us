### Education data 

source: 

```python
ed=pd.read_csv('../csv_s/ed_data_voting_age.csv')
ed=ed.rename(columns={'Total_Voting_Age_Pop:': 'vote_age_pop','Graduate_or_professional_degree': 'grad_prof_deg'})
```

```python
#Group & Find Pct of Population & Rank
#no hs diploma is ranked highest % of no degree as lowest
#no diploma at all
ed['no_diploma']= ed[['<_9th_grade', '9th_to_12th_grade_no diploma']].sum(axis=1)
ed['no_diploma_pct']= ed['no_diploma']/ed['vote_age_pop']*100
ed['no_diploma_rank'] = ed['no_diploma_pct'].rank(method='max', na_option='bottom').astype(int)

#rest are ranked with highest % with degrees as highest
#hs degree but no assoc or higher
ed['hs_nodegree']= ed[['High_school_graduate_(includes_equivalency)', 'Some_college_no_degree']].sum(axis=1)
ed['hs_nodegree_pct']=ed['hs_nodegree']/ed['vote_age_pop']*100
ed['hs_nodegree_rank'] = ed['hs_nodegree_pct'].rank(ascending=False, method='max', na_option='bottom').astype(int)

#2 or 4 year degree
ed['assoc_bach']= ed[['Associates_degree', 'Bachelors_degree']].sum(axis=1)
ed['assoc_bach_pct']=ed['assoc_bach']/ed['vote_age_pop']*100
ed['assoc_bach_rank'] = ed['assoc_bach_pct'].rank(ascending=False, method='max', na_option='bottom').astype(int)

# >4 year degree
ed['grad_prof_pct']=ed['grad_prof_deg']/ed['vote_age_pop']*100
ed['grad_prof_rank'] = ed['grad_prof_pct'].rank(ascending=False, method='max', na_option='bottom').astype(int)

```

```python
#Drop Grouped Columns
ed=ed.drop(columns=['9th_to_12th_grade_no diploma','<_9th_grade', 
                    'High_school_graduate_(includes_equivalency)',
                    'Some_college_no_degree','Associates_degree','Bachelors_degree'])
```

```python
# create a total of where each state state ranked and rank that column for total ed ranking 
# decided not to weight these but maybe should have?
ed['ed_total']= ed['no_diploma_rank']+ed['hs_nodegree_rank']+ed['assoc_bach_rank']+ed['grad_prof_rank']
ed['overall_ed_rank'] = ed['ed_total'].rank(method='max', na_option='bottom').astype(int)
```

### Teen Birth Data

source: https://www.cdc.gov/nchs/pressroom/sosmap/teen-births/teenbirths.htm

```python
birth=pd.read_csv('../csv_s/teen_birth_rate.csv')
birth.head()
```

```python
#narrow to 2020 &  drop unnecessary columns
birth=birth[birth['YEAR']==2020].drop(columns={'YEAR', 'URL'}).rename(columns={'RATE': 'teen_birth_rate'})
birth.head()
```

```python
#replace two letter code with full state name for big merge
# do it now to add DC with NaN
state= pd.read_csv('csv_s/state_abbv.csv')
births=birth.merge(state, how='outer', left_on='STATE', right_on='Abbv').drop(columns={"STATE", "Abbv"})
births.head()
```

```python
births['teen_birth_rank'] =births['teen_birth_rate'].rank(method='min', na_option='bottom').astype(int)
births.head()
```

### Merge Education & Births

```python
edbirth=ed.merge(births, how="outer", left_on='State', right_on='State')
```

```python
edbirth.head()
```

### 2020 Voting 

source: https://www.census.gov/data/tables/time-series/demo/voting-and-registration/p20-585.html (table04a)

```python
voting= pd.read_csv('../csv_s/2020_voting.csv', usecols=['State','Total_registered','Percent_registered','Total_voted', 'Percent_voted'])
voting.head()
```

```python
# Convert to int from object
voting['Total_registered']= voting['Total_registered'].str.replace(',', '').astype(int)
voting['Total_voted']= voting['Total_voted'].str.replace(',', '').astype(int)
```

```python
# registered and voted were per 1000, multiply to keep consistent with rest of dataset
voting["Total_registered"] = 1000 * voting["Total_registered"]
voting["Total_voted"] = 1000 * voting["Total_voted"]
```

```python
#assign rank
voting['voted_rank'] =voting['Percent_voted'].rank(ascending=False, method='max', na_option='bottom').astype(int)
voting['registered_rank'] =voting['Percent_registered'].rank(ascending=False, method='max', na_option='bottom').astype(int)
```

```python
#Change State from all caps to proper caps for merge
voting["State"]=voting['State'].str.title()
# Beacuse title capitalizes all first letters, DC has to be handled uniquely for consistency
voting.at[8,'State']='District of Columbia'
```

```python
voting
```

### Merge Voting with Education & Birth

```python
ed_birth_vote=edbirth.merge(voting, how='outer', left_on='State', right_on='State').fillna(0)
```
