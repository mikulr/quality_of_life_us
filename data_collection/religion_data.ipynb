```python
import pandas as pd
import numpy as np
from selenium import webdriver
from bs4 import BeautifulSoup
from webdriver_manager.chrome import ChromeDriverManager
```

```python
driver = webdriver.Chrome(ChromeDriverManager().install())
driver.get('https://www.pewresearch.org/fact-tank/2016/02/29/how-religious-is-your-state/?state=alabama')
```

```python
# default table is states ranked by overall combined index
soup = BeautifulSoup(driver.page_source, 'lxml')
#find table on page
tables = soup.find_all('table')
#read table
overall = pd.read_html(str(tables))
# put table in dataframe to work with data
overall_df=pd.DataFrame(np.concatenate (overall))
overall_df.head()
```

```python
#take column 0 in df split it to get name and ranking into new df
new_over= overall_df[0].str.split('.', expand=True).rename(columns={1: 'State' , 0:'Rel_Overall_Rank'})
new_over.head()
```

```python
# merge two df's together at index taking only column 1 from original table
final_over = new_over.merge(overall_df.iloc[:,[1]], how='left', left_index=True, right_index=True)
final_over = final_over.rename(columns={1: 'Rel_Overall_Pct'})
final_over['Rel_Overall_Pct']=final_over['Rel_Overall_Pct'].str.strip("%")
final_over.head()
```

```python
#manually change dropdown on page to select new table to worship attendance
#format will be same as above
soup = BeautifulSoup(driver.page_source, 'lxml')
tables = soup.find_all('table')
attendance = pd.read_html(str(tables))
attendance_df=pd.DataFrame(np.concatenate (attendance))
attendance_df.head()
```

```python
new_att= attendance_df[0].str.split('.', expand=True).rename(columns={0:'Rel_Att_Rank', 1: "State"})
new_att.head()
```

```python
final_attendance=new_att.merge(attendance_df.iloc[:,[1]], how='left', left_index=True, right_index=True)
final_attendance = final_attendance.rename(columns={1: 'Rel_Att_Pct'})
final_attendance['Rel_Att_Pct']=final_attendance['Rel_Att_Pct'].str.strip("%")
final_attendance.head()
```

```python
#manually change dropdown on page to select new table to belief in God
soup = BeautifulSoup(driver.page_source, 'lxml')
tables = soup.find_all('table')
belief = pd.read_html(str(tables))
belief_df=pd.DataFrame(np.concatenate (belief))
# belief_df.head()
```

```python
new_belief= belief_df[0].str.split('.', expand=True).rename(columns={0:'Rel_Belief_Rank', 1: "State"})
# new_belief.head()
```

```python
final_belief=new_belief.merge(belief_df.iloc[:,[1]], how='left', left_index=True, right_index=True)
final_belief = final_belief.rename(columns={1: 'Rel_Belief_Pct'})
final_belief['Rel_Belief_Pct']=final_belief['Rel_Belief_Pct'].str.strip("%")
final_belief.head()
```

```python
#manually change dropdown on page to select new table to importance of religion
soup = BeautifulSoup(driver.page_source, 'lxml')
tables = soup.find_all('table')
importance = pd.read_html(str(tables))
import_df=pd.DataFrame(np.concatenate (importance))
# import_df
```

```python
new_import= import_df[0].str.split('.', expand=True).rename(columns={0:'Rel_Import_Rank', 1: "State"})
```

```python
final_import=new_import.merge(import_df.iloc[:,[1]], how='left', left_index=True, right_index=True)
final_import = final_import.rename(columns={1: 'Rel_Import_Pct'})
final_import['Rel_Import_Pct']=final_import['Rel_Import_Pct'].str.strip("%")
final_import.head()
```

```python
#manually change dropdown on page to select new table to freq of prayer
soup = BeautifulSoup(driver.page_source, 'lxml')
tables = soup.find_all('table')
freq = pd.read_html(str(tables))
freq_df=pd.DataFrame(np.concatenate (freq))
freq_df.head()
```

```python
new_freq= freq_df[0].str.split('.', expand=True).rename(columns={0:'Rel_Prayer_Freq_Rank', 1: "State"})
# new_freq.head()
```

```python
final_freq = new_freq.merge(freq_df.iloc[:,[1]], how='left', left_index=True, right_index=True)
final_freq = final_freq.rename(columns={1: 'Rel_Freq_Prayer_Pct'})
final_freq['Rel_Freq_Prayer_Pct']=final_freq['Rel_Freq_Prayer_Pct'].str.strip("%")
final_freq.head()
```

```python
merge1=final_import.merge(final_belief, how='inner', left_on='State', right_on='State')
merge2=final_attendance.merge(final_over, how='inner', left_on='State', right_on='State')
```

```python
religion1=merge1.merge(merge2, how='inner', left_on='State', right_on='State')
religion=religion1.merge(final_freq, how='inner', left_on='State', right_on='State')
religion.head()
```

```python
religion= religion[['State','Rel_Overall_Rank','Rel_Overall_Pct','Rel_Import_Rank','Rel_Import_Pct','Rel_Belief_Rank','Rel_Belief_Pct','Rel_Att_Rank',
'Rel_Att_Pct', 'Rel_Prayer_Freq_Rank', 'Rel_Freq_Prayer_Pct']]
```

```python
#fix formatting and type
religion['State']=religion['State'].str.lstrip()
numeric=religion.columns.drop('State')
religion[numeric]=religion[numeric].apply(pd.to_numeric)
religion.dtypes
```

```python
religion.to_json('../json_out/religion.json', orient='records')
```
