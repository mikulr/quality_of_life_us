```python
import pandas as pd
from bs4 import BeautifulSoup as bs
import requests
```

```python
 pop= pd.read_html('https://en.wikipedia.org/wiki/List_of_U.S._states_and_territories_by_population')
```

```python
#take first table on page
pop=pop[0]
pop.head()
```

```python
#grab population numbers
pop2= pop[['Census population[8][a]'][0]]
pop3=pop2['July 1, 2021']
```

```python
#grab state names
st=pop[['State or territory'][0]]
st.head()
```

```python
# create new dF with just name and popultaion
population= pd.concat([st,pop3] , axis=1).rename(columns={'State or territory':"State", 'July 1, 2021':"total_population"})
population
```

```python
#delete territories and totals 
#Puerto Rico = 29
population=population.drop(range(52,60))
population = population.drop([29])
```

```python
population['pop_rank']=population['total_population'].rank(ascending=False, method='min', na_option='bottom').astype(int)
population
```

### Unemployment data for May 2022 

```python
empl= pd.read_html('https://www.bls.gov/web/laus/laumstrk.htm')
#empl is several tables, put the first one into a dataframe
emp_df = pd.DataFrame(empl[0])
emp_df = emp_df.rename(columns={'May 2022(p)rate': 'may_22_unemp', "Rank": 'unemp_rank'})
#drop out the footer from scraped table
emp_df= emp_df.drop(range(51,54))
emp_df.head()
```

```python
# convert select columns to numeric 
emp_df[['may_22_unemp', 'unemp_rank']]= emp_df[['may_22_unemp', 'unemp_rank']].apply(pd.to_numeric)
```

```python
# Merge to population data
popemp=population.merge(emp_df, left_on='State', right_on="State")
popemp.dtypes
```

```python
popemp.to_json('../json_out/popemp.json', orient='records')
```
